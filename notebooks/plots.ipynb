```python
json_path = '/midtier/sablab/scratch/alw4013/keymorph/experiments/__eval__gigamed-keymorph_keypoints256_batch1_normTypeinstance_lr3e-06/results/summary.json'

import json

with open(json_path) as f:
    d = json.load(f)
    
all_aff = []
all_rig = []
all_tps = []
for key in d.keys():

    if 'affine' in key:
        all_aff += d[key]
    if 'rigid' in key:
        all_rig += d[key]
    if 'tps' in key:
        all_tps += d[key]
        
import seaborn as sns
sns.set_theme()
sns.boxplot({'aff':all_aff, 'rig':all_rig, 'tps':all_tps})
```
